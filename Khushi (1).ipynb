```python
import pandas as pd

# Load dataset (with the correct CSV file name)
df = pd.read_csv("22yztw (2).csv")


# Print the DataFrame
print(df)
print(df.isnull().sum())
# Replace numeric column NULLs with mean
for col in df.select_dtypes(include='number').columns:
    df[col].fillna(df[col].mean(), inplace=True)
def sugar_condition(value):
    if value < 70:
        return "Low"
    elif value <= 99:
        return "Normal"
    elif value <= 125:
        return "Prediabetes"
    else:
        return "Diabetes"

df["Sugar_Condition"] = df["sugar_level"].apply(sugar_condition)
print(df)
# Remove duplicate rows
df = df.drop_duplicates()
print(df)
# Calculate average visit cost
avg_cost = df["visit_cost"].mean()

print("Average Visit Cost:", avg_cost)
# Filter diabetes patients
diabetes_df = df[df["Sugar_Condition"] == "Diabetes"]

# Calculate average sugar level
avg_sugar = diabetes_df["sugar_level"].mean()

print("Average Sugar Level for Diabetes:", avg_sugar)
```
